### Prepare Library

```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

### Prepare Image

Source : https://www.census.gov/history/img/einstein-laurencelivermorenl.jpg

```
image = cv2.imread('einstein.jpg', cv2.IMREAD_UNCHANGED)
```

### Create Helper Function


```
def showImage(real, filtered, title_real, title_filter):
  plt.figure(figsize=(15,15))
  plt.subplot(121),plt.imshow(real),plt.title(title_real)
  plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(filtered),plt.title(title_filter)
  plt.xticks([]), plt.yticks([])
  plt.show()
```

## Number 1
#### (Image Blurring/Image Smoothing)

Refer to documentation : https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_filtering/py_filtering.html#image-blurring-image-smoothing

Number 1

```
blur = cv2.blur(image,(3,3)) #kernel 3x3
showImage(image, blur, "Original", "Filtered")
```

## Number 2
#### (Linear Filters : Shift Left 1 px)

Refer to documentation : https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_filtering/py_filtering.html#d-convolution-image-filtering

Number 2

```
kernel = np.array([[0, 0, 0],
                   [0, 0, 1],
                   [0, 0, 0]])
filtered = cv2.filter2D(image,-1,kernel)

showImage(image, filtered, 'Original', 'Shifted')
```

## Number 3
#### (Linear Filters)

Refer to documentation : https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_filtering/py_filtering.html#d-convolution-image-filtering

Number 3

```
kernel = np.array([[0, 0, 0],
                   [0, 2, 0],
                   [0, 0, 0]])
filtered = cv2.filter2D(image,-1,kernel)

showImage(image, filtered, 'Original', 'Filtered')
```

## Number 4
#### (Linear Filters)

Refer to documentation : https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_filtering/py_filtering.html#d-convolution-image-filtering

Number 4

```
kernel = np.ones((3, 3), np.float32) / 3
filtered = cv2.filter2D(image,-1,kernel)

showImage(image, filtered, 'Original', 'Filtered')
```

## Number 5
#### (Linear Filters : Sharpening)

Refer to documentation : https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_filtering/py_filtering.html#d-convolution-image-filtering

Number 5

```
first_array = np.array([[0, 0, 0],
                        [0, 2, 0],
                        [0, 0, 0]])
second_array = np.ones((3, 3), np.float32) / 9
kernel = first_array - second_array
filtered = cv2.filter2D(image,-1,kernel)

showImage(image, filtered, 'Original', 'Sharpened')
```

## Number 6
#### (Linear Filters : Gaussian Blur)

Refer to documentation : https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_filtering/py_filtering.html#gaussian-filtering

Number 6

```
gaussian = cv2.GaussianBlur(image,(5,5),0) #kernel 5x5
showImage(image, gaussian, "Original", "Gaussian Filtered")
```
